```python
# ==========================================
# Mencari Hasil K-Means Clustering - Café Komunikoffie
# ==========================================

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load data
df = pd.read_excel("Data_Penjualan_Rapi_Mentah.xlsx")

# Pilih fitur numerik yang relevan
X = df[['Total Penjualan', 'Jumlah Terjual', 'Laba', 'Jumlah Order']]

# Normalisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Tentukan jumlah cluster (misal 3)
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Urutkan cluster berdasarkan rata-rata total penjualan
cluster_order = df.groupby('cluster')['Total Penjualan'].mean().sort_values().index

cluster_labels = {
    cluster_order[0]: "Penjualan Rendah",
    cluster_order[1]: "Penjualan Stabil",
    cluster_order[2]: "Penjualan Tinggi"
}

# Tambah kolom label hasil interpretasi
df['label_cluster'] = df['cluster'].map(cluster_labels)

# Simpan hasil
df.to_excel("Hasil_Clustering_Penjualan.xlsx", index=False)

print("=== Contoh Hasil Cluster ===")
print(df[['Produk','Total Penjualan','cluster','label_cluster']].head())

print("\nJumlah per cluster:")
print(df['label_cluster'].value_counts())
```
